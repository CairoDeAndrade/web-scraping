Agora vamos adicionar o **"yield"**, que  parecido com um **return** porm **return** retorna um valor especificado, enquanto Yield pode produzir uma sequncia de valores.

```python
import scrapy


class BistekSpider(scrapy.Spider):
    nome = 'bistek'
    start_urls = ['https://www.bistek.com.br/bebidas/vinhos-e-espumantes.html']

     def parse(self, response):
      for bistek in response.css('.product-item-info'):
        nome = response.css('.product-item-link::text').get().replace('\n', '').replace(' ', '')
        preo = response.css('.price::text').get().split()[1]
        link = response.xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "product-item-link", " " ))]//@href').get()
          yield {
                'preo': preo,
                'nome': nome,
                'link': link
            }

      
```

Nosso cdigo j consegue pegar todas os dados que queremos de uma pgina, mas, agora vamos fazer uma extrao em **todas** as pginas existente neste site.

Para isso, vamos usar o **xpath** para achar onde nos pulamos para prxima pgina e coletar esse elemento.

exemplo:

image.png

Coletando isso, vamos agora us-lo para, assim que a raspagem da primeira pgina inteira terinar, nos pularmos para prxima, at acabar.

```python
import scrapy


class BistekSpider(scrapy.Spider):
    nome = 'bistek'
    start_urls = ['https://www.bistek.com.br/bebidas/vinhos-e-espumantes.html']

     def parse(self, response):
      for bistek in response.css('.product-item-info'):
        nome = response.css('.product-item-link::text').get().replace('\n', '').replace(' ', '')
        preo = response.css('.price::text').get().split()[1]
        link = response.xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "product-item-link", " " ))]//@href').get()
          yield {
                'preo': preo,
                'nome': nome,
                'link': link
            }

      prox_pag = response.xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "next", " " ))]//@href').get()
      if prox_pag is not None:
        yield response.follow(prox_pag, callback=self.parse)
```

**ORDENADO OS PREOS**

Agora j temos o cdigo de extrao pronto, prm iremos ordenar todas estas informaes por preo, em forma crescente, para pode visualizar os preos menores at os maiores de forma mais visvel. (isso  um exemplo do que podeiria uma emprsa fazer com os preos dos concorrentes, afim de precificar seu prprio produto)

**Cdigo**

Para fazer essa ordenao, utilizamos um **"for"** para varrer todos os preos e um **"sorted()"** para orden-los. (lembrando que este cdigo para ordenar **DEVE** ser feito fora da spider(programa da raspagem) para evitar erros, ento voc pode criar um novo arquivo no seu projeto e faz-lo)

```python
import json
from operator import itemgetter

with open('bistek.json') as bistek:
    bistek_data = json.load(bistek)

    for i, j in enumerate(bistek_data):
        bistek_data[i]['price'] = float(bistek_data[i]['price'].replace(',', '.'))

    sorted_price1 = sorted(bistek_data, key=itemgetter('price'), reverse=False)

    for i in sorted_price1:
        print(i)
```
