**SCRAPY FRAMEWORK**


**O que são Web crawlers?**
  Antes de estudarmos o scrapy, é primordial que saibamos o que é um **Web crawler**. Nessa perspectiva, resumidamente, um Web crawler é um programa usado para **colher conteúdo** de um website em busca de informações úteis, através do protocolo padrão da web (http/https), que são protocolos de comunicação que permitem a **troca de dados** entre o seu computador ou dispositivo móvel e um servidor que abriga um determinado site que você acessa.




**Scrapy, o que é?**

  Scrapy é um **Framework**, em outras palavras, uma espécie de template, um conjunto de ferramentas que contam com diversas funcionalidades, e, além do mais, é **open source**, ou seja, pode ser visualizado, modificado e distribuído por qualquer pessoa.   O Scrapy é utilizado para **extração** de informação em websites, isto é, Framework para Web Crawler.

Consequentemente, Web crawlers, como o scrapy no python, são bastante utilizados para diversos objetivos, como:

 Motores de busca

Agora vamos adicionar o **"yield"**, que é parecido com um **return** porém **return** retorna um valor especificado, enquanto Yield pode produzir uma sequência de valores.

```python
import scrapy


class BistekSpider(scrapy.Spider):
    nome = 'bistek'
    start_urls = ['https://www.bistek.com.br/bebidas/vinhos-e-espumantes.html']

     def parse(self, response):
      for bistek in response.css('.product-item-info'):
        nome = response.css('.product-item-link::text').get().replace('\n', '').replace(' ', '')
        preço = response.css('.price::text').get().split()[1]
        link = response.xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "product-item-link", " " ))]//@href').get()
          yield {
                'preço': preço,
                'nome': nome,
                'link': link
            }

      
```

Nosso código já consegue pegar todas os dados que queremos de uma página, mas, agora vamos fazer uma extração em **todas** as páginas existente neste site.

Para isso, vamos usar o **xpath** para achar onde nos pulamos para próxima página e coletar esse elemento.

exemplo:

image.png

Coletando isso, vamos agora usá-lo para, assim que a raspagem da primeira página inteira terinar, nos pularmos para próxima, até acabar.

```python
import scrapy


class BistekSpider(scrapy.Spider):
    nome = 'bistek'
    start_urls = ['https://www.bistek.com.br/bebidas/vinhos-e-espumantes.html']

     def parse(self, response):
      for bistek in response.css('.product-item-info'):
        nome = response.css('.product-item-link::text').get().replace('\n', '').replace(' ', '')
        preço = response.css('.price::text').get().split()[1]
        link = response.xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "product-item-link", " " ))]//@href').get()
          yield {
                'preço': preço,
                'nome': nome,
                'link': link
            }

      prox_pag = response.xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "next", " " ))]//@href').get()
      if prox_pag is not None:
        yield response.follow(prox_pag, callback=self.parse)
```

**ORDENADO OS PREÇOS**

Agora já temos o código de extração pronto, prém iremos ordenar todas estas informações por preço, em forma crescente, para pode visualizar os preços menores até os maiores de forma mais visível. (isso é um exemplo do que podeiria uma emprsa fazer com os preços dos concorrentes, afim de precificar seu próprio produto)

**Código**

Para fazer essa ordenação, utilizamos um **"for"** para varrer todos os preços e um **"sorted()"** para ordená-los. (lembrando que este código para ordenar **DEVE** ser feito fora da spider(programa da raspagem) para evitar erros, então você pode criar um novo arquivo no seu projeto e fazê-lo)

```python
import json
from operator import itemgetter

with open('bistek.json') as bistek:
    bistek_data = json.load(bistek)

    for i, j in enumerate(bistek_data):
        bistek_data[i]['price'] = float(bistek_data[i]['price'].replace(',', '.'))

    sorted_price1 = sorted(bistek_data, key=itemgetter('price'), reverse=False)

    for i in sorted_price1:
        print(i)
```
